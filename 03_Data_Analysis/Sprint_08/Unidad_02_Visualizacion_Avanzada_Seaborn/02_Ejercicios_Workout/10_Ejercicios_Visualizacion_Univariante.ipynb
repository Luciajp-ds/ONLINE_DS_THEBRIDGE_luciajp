### Ejercicio 2

Crea una funcin a la que puedas pasar como argumentos un dataframe y una lista de columnas categricas y dibuje una matriz de grficas de 2 columnas y tantas filas como necesite para pintar el diagrama de barras de las frecuencias absolutas de las columnas pasadas como argumento. No hace falta que compruebes que las columnas son categricas. Y que ponga como ttulo de cada figura "Frecuencias Absolutas de \<nombre de la columna>\. Utiliza Seaborn para dibujar las grficas y matplotlib para lo que necesites (pero no para dibujar las grficas)

```python
def barras_categoricas(df, lista_col_cat):
    num_filas = round(len(lista_col_cat)/2)
    fig,axs = plt.subplots(nrows=num_filas,ncols=2, figsize=(14,5*num_filas))

    axs = axs.ravel()       # para indexar linealmente y facilitar el bucle
    for i, col in enumerate(lista_col_cat):
        sns.countplot(x=col, data = df, ax=axs[i])
        axs[i].set_title(f"Frecuencias absolutas de {col}", fontsize=14)
    
    #oculto eje vaco si son un nmero impar de columnas
    for j in range(i+1, len(axs)):
        axs[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```

### Ejercicio 4

Crea una funcin similar a la del ejercicio 4, pero que esta vez pinte el histograma de una lista de columnas numricas, aade un argumento para poder indicar el nmero de bins. Aade tambin un argumento que permita incluir la funcin de densidad o no en la figura. De nuevo usa Seaborn para crear las figuras. Tambin debe cumplirse el que pinte cada grafica en una matriz de 2 columnas como mucho y todas las filas que necesites.

```python
def histo_numericas(df, lista_col_num, num_bins=30, fun_densidad=False):
    num_filas = round(len(lista_col_num)/2)
    fig,axs = plt.subplots(nrows=num_filas,ncols=2, figsize=(14,5*num_filas))

    axs = axs.ravel()       # para indexar linealmente y facilitar el bucle
    for i, col in enumerate(lista_col_num):
        sns.histplot(x= col, data = df, kde= fun_densidad, bins= num_bins, ax = axs[i])
        axs[i].set_title(f"Histograma de {col}", fontsize=14)

    
    #oculto eje vaco si son un nmero impar de columnas
    for j in range(i+1, len(axs)):
        axs[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```

### Ejercicio 6

Construye una funcin a la que puedas pasar un dataframe y una serie de columnas numricas (que suponemos son series temporales o secuencias) y que pinte cada columna en una sola grfica (de nuevo en una matriz de 2 columnas como mucho).

```python
def plot_series_numericas(df, lista_col_num):
    num_filas = round(len(lista_col_num)/2)
    fig,axs = plt.subplots(nrows=num_filas,ncols=2, figsize=(14,5*num_filas))

    axs = axs.ravel()       # para indexar linealmente y facilitar el bucle
    for i, col in enumerate(lista_col_num):
        axs[i].plot(df[col], linewidth=2)
        axs[i].set_title(f"Serie temporal de {col}", fontsize=14)
        axs[i].set_xlabel("ndice")

    
    #oculto eje vaco si son un nmero impar de columnas
    for j in range(i+1, len(axs)):
        axs[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```

### Ejercicio 8

Aade tres argumentos a la funcion del ejercicio 6. El primero debe permitir pintar una sola grfica con todas las lneas de evolucin juntas y aadir una leyenda. El segundo debe permitir colocar una fecha de inicio y el tercero una fecha de fin. Si no se dan valores a estos argumentos la figura debe contemplarlos todos, si se dan la figura slo debe mostrar los valores comprendidos entre esas fechas. Supon que la fecha es el ndice del DataFrame. De nuevo las grficas deben estar pintadas con Seaborn.

```python
def plot_series_numericas_v2(df, lista_col_num, todas_en_una=False, inicio=None, fin=None):
    
    # --- Filtrado temporal, el ndice del df tiene que ser datetime ---
    if inicio is not None or fin is not None:
        df_filtrado = df.loc[inicio:fin]
    else:
        df_filtrado = df.copy()

    # --- CASO 1: Todas las series en una sola grfica ---
    if todas_en_una:
        plt.figure(figsize=(14, 6))

        for col in lista_col_num:
            sns.lineplot(data=df_filtrado, x=df_filtrado.index, y=col, linewidth=2, label=col)

        plt.title("Series temporales agrupadas")
        plt.xlabel("Fecha")
        plt.ylabel("Valor")
        plt.legend()
        plt.tight_layout()
        plt.show()
        return  # Salimos porque no hay subplots
    
    # --- CASO 2: Subplots individuales ---
    num_filas = round(len(lista_col_num)/2)
    fig, axs = plt.subplots(nrows=num_filas, ncols=2, figsize=(14, 5 * num_filas))
    axs = axs.ravel()

    for i, col in enumerate(lista_col_num):
        sns.lineplot(ax=axs[i], data=df_filtrado, x=df_filtrado.index, y=col, linewidth=2)
        axs[i].set_title(f"Serie temporal de {col}", fontsize=14)
        axs[i].set_xlabel("Fecha")

    # Ocultar ejes vacos si nmero impar de columnas
    for j in range(i + 1, len(axs)):
        axs[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```
