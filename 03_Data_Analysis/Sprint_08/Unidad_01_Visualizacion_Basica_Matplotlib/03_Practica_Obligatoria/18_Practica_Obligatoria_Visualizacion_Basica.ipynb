#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

fig.suptitle("Resumen de Grficos - Ciudades de California", fontsize=16)

# ======================
# 1. Ciudades ms pobladas
# ======================
ax = axs[0, 0]
top_cities = cities.nlargest(5, "population_total")

bars = ax.bar(
    top_cities["city"],
    top_cities["population_total"],
    color=["blue", "grey", "red", "olive", "brown"],
    alpha=0.7
)
for i, (city, pop) in enumerate(zip(top_cities["city"], top_cities["population_total"])):
    ax.text(i, pop + (pop * 0.02), f"{int(pop):,}", ha="center")
ax.set_title("Ciudades ms pobladas")
ax.set_ylabel("Poblacin")

# ======================
# 2. Histograma rea total
# ======================
ax = axs[0, 1]
ax.hist(cities["area_total_km2"], bins=50, color="steelblue")
ax.set_title("Histograma del rea total")
ax.set_xlabel("Superficie (km$^2$)")
ax.set_ylabel("Frecuencia")
ax.legend()

# ======================
# 3. Scatter agua vs poblacin
# ======================
ax = axs[1, 0]
ax.scatter(cities["area_water_km2"], cities["population_total"], alpha=0.6, color="green")
ax.set_title("rea de agua vs Poblacin")
ax.set_xlabel("rea de agua")
ax.set_ylabel("Poblacin")

# Outliers
x = cities["area_water_km2"]
y = cities["population_total"]

# Calculamos IQR de cada variable
Q1_x, Q3_x = x.quantile(0.25), x.quantile(0.75)
IQR_x = Q3_x - Q1_x

Q1_y, Q3_y = y.quantile(0.25), y.quantile(0.75)
IQR_y = Q3_y - Q1_y

# Umbral (puede ser 1.5 o 3)
k = 100

# Condicin de outliers
outliers = cities[
    ((x < Q1_x - k*IQR_x) | (x > Q3_x + k*IQR_x)) |
    ((y < Q1_y - k*IQR_y) | (y > Q3_y + k*IQR_y))
]

outliers[["city", "area_water_km2", "population_total"]]


props_flecha = {
    "arrowstyle" : "<->",
    "connectionstyle" : "angle3"
}

#for i in range(len(outliers)):
#    ax.annotate("Outliers", xy=(outliers["area_water_km2"][i], outliers["population_total"][i]), xytext=(0.5, 0.5), arrowprops=props_flecha)


# ======================
# 4. Boxplot elevacin
# ======================
ax = axs[1, 1]
ax.boxplot(
    cities["elevation_m"],
    patch_artist=True,
    boxprops=dict(facecolor="red", color="black"),
    medianprops=dict(color="black"),
    vert = False
)
ax.set_title("Boxplot de elevacin")
ax.set_ylabel("Elevacin")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
